```python
import os
import gradio as gr
# 🔴 Comment this (OpenAI import)
# from langchain.chat_models import ChatOpenAI
# ✅ Instead, import HuggingFace
from langchain_community.llms import HuggingFaceHub
from langchain import LLMChain, PromptTemplate
from langchain.memory import ConversationBufferMemory
from transformers import pipeline

from langchain.llms import HuggingFacePipeline
```

```python
template = """You’re an alien disguised as a human, sent to Earth to study human behavior. You observe everyday life with childlike wonder and ask strange but insightful questions about human customs, emotions, and pizza.
{chat_history}
User: {user_message}
Chatbot:"""

prompt = PromptTemplate(
    input_variables=["chat_history", "user_message"],
    template=template
)

memory = ConversationBufferMemory(memory_key="chat_history")
```

```python
hf_pipeline = pipeline(
    "text-generation",
    model="tiiuae/falcon-7b-instruct",  # You can change this model
    max_length=512,
    temperature=0.5
)
llm = HuggingFacePipeline(pipeline=hf_pipeline)


llm_chain = LLMChain(
    llm=llm,
    prompt=prompt,
    verbose=True,
    memory=memory,
)
```

```python
HUGGINGFACEHUB_API_TOKEN = "paste_your_token_here"
os.environ["HUGGINGFACEHUB_API_TOKEN"] = HUGGINGFACEHUB_API_TOKEN
```

```python
def get_text_response(user_message, history):
    response = llm_chain.predict(user_message=user_message)
    return response
```

```python
demo = gr.ChatInterface(get_text_response, examples=["How are you doing?","What are your interests?","Which places do you like to visit?"])
```

```python
if __name__ == "__main__":
    demo.launch(debug=True)
```
